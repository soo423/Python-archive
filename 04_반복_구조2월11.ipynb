```python
emplovee_num = int(input('ì§ì› ìˆ˜ ì…ë ¥'))
total = 0

for i in range(1, emplovee_num+1):
  print(f'{i}ë²ˆì§¸ ì§ì› ì •ë³´ ì…ë ¥')
  rank = int(input('ì§ê¸‰ ì…ë ¥'))
  basic_salary = int(input('ë³¸ë´‰ ì…ë ¥'))

if rank >= 2:
  bonus = basic_salary * 0.3
  salary = basic_salary + bonus
  print(f'2ê¸‰ ì´ìƒ ë³´ë„ˆìŠ¤ë¡œ ì§€ê¸‰ëœ ê¸ˆì•¡ì€ {bonus}ë§Œì› ì…ë‹ˆë‹¤.')
else:
  bonus = basic_salary * 0.2
  salary = basic_salary + bonus
  print(f'2ê¸‰ ë¯¸ë§Œ ë³´ë„ˆìŠ¤ë¡œ ì§€ê¸‰ëœ ê¸ˆì•¡ì€ {bonus}ë§Œì› ì…ë‹ˆë‹¤.')

total += salary
print(f'{i}ë²ˆì§¸ ì§ì›ì˜ ì´ ì›”ê¸‰ì€ {total}')
```

### ë¬¸ì œ : ê²½ì§„ëŒ€íšŒ í‰ê°€ ì ìˆ˜ êµ¬í•˜ê¸°

*   ì‹¬ì‚¬ìœ„ì› ë³€ìˆ˜ : jugde_numì€ 5ëª…ì„
*   ì ìˆ˜ì˜ ì´ í•©ê³„ sumì˜ í‰ê· ì„ ì¶œë ¥
*   ìµœê³  ì ìˆ˜ëŠ” ë¹¼ê³  í•©ê³„ë¥¼ êµ¬í•¨ max_score
*   í˜„ì¬ ì ìˆ˜ê°€ ìµœê³  ì ìˆ˜ë³´ë‹¤ í¬ë©´, í˜„ì¬ ì ìˆ˜ë¥¼ max_scoreì— ì €ì¥, sumì—ì„œ ë¹¼ì•¼í•¨


```python
jugde_num = 5
sum = 0
max_score = 0

for i in range(1, jugde_num+1):
  score = int(input(f'{i}ë²ˆì´ ì‹¬ì‚¬í•œ ì ìˆ˜ ì…ë ¥'))
  sum += score

  if score > max_score:
    max_score = score
    sum -= max_score

average = sum / (jugde_num-1)

print(f'ì´í•© ì ìˆ˜ : {sum+jugde_num}')
print(f'ì‹¬ì‚¬ ê²°ê³¼ ìµœê³  ì ìˆ˜ : {max_score}')
print(f'í‰ê·  ì ìˆ˜ {average}')


```

### ğŸ¯ ìˆ«ì ë§íˆê¸° ê²Œì„
1. **ì»´í“¨í„°ê°€ ìƒê°í•œ ìˆ˜ ë§íˆê¸°**:
    * `random.randint(1, 100)`ìœ¼ë¡œ ì •ë‹µì„ ìƒì„±í•©ë‹ˆë‹¤.
    * ì‚¬ìš©ìê°€ ì •ë‹µì„ ë§í ë•Œê¹Œì§€ `while` ë¬¸ìœ¼ë¡œ ì…ë ¥ì„ ë°˜ë³µí•©ë‹ˆë‹¤.
    * ì…ë ¥ê°’ì´ í¬ë©´ "ë„ˆë¬´ ë†’ì•„ìš”", ì‘ìœ¼ë©´ "ë„ˆë¬´ ë‚®ì•„ìš”"ë¥¼ ì¶œë ¥í•˜ì—¬ ë²”ìœ„ë¥¼ ì¢í˜€ê°‘ë‹ˆë‹¤.
2. **ì‚¬ìš©ìê°€ ìƒê°í•œ ìˆ˜ ë§íˆê¸°**:
    * ì‚¬ìš©ìê°€ ìˆ«ìë¥¼ í•˜ë‚˜ ì •í•˜ë©´ ì»´í“¨í„°ê°€ `low`ì™€ `high` ë³€ìˆ˜ë¥¼ ì¡°ì •í•˜ë©° ì •ë‹µì„ ì°¾ì•„ê°€ëŠ” ë¡œì§ì…ë‹ˆë‹¤.

```python
#1. ì»´í“¨í„°ê°€ ìƒê°í•œ ìˆ«ì ë§ì¶”ê¸°
import random

print('1ë¶€í„° 100ê¹Œì§€ì˜ ìˆ«ì í•˜ë‚˜ë¥¼ ì •í–ˆìŠµë‹ˆë‹¤.')
print('ì´ ìˆ«ìë¥¼ ë§ì¶°ë³´ì„¸ìš”')

answer = random.randint(1,100) # 1~100ìˆ«ì í•˜ë‚˜ ëœë¤ìœ¼ë¡œ ì§€ì •ã…Œ
time = 0

#whileë¬¸ìœ¼ë¡œ ë°˜ë³µí•˜ëŠ”ë°
#userê°€ ì…ë ¥í•œ ìˆ«ìê°€ ë§ìœ¼ë©´ ë°˜ë³µë¬¸ ì¢…ë£Œ
#userê°€ ì…ë ¥í•œ ìˆ«ìê°€ í¬ë©´ í¬ë‹¤ê³  ì¶œë ¥í•¨
#userê°€ ì…ë ¥í•œ ìˆ«ìê°€ ì‘ìœ¼ë©´ ì‘ë‹¤ê³  ì¶œë ¥í•¨
#ìµœì¢… ë„ì „ ë° í™”ë©´ì— ì •ë‹µì„ ë§ì·„ëŠ”ì§€ ì¶œë ¥í•¨

while True:
  user = int(input('ìˆ«ì ì…ë ¥'))
  time += 1
  if user == answer :
    break

  elif answer > user:
    print("ì»´í“¨í„°ê°€ ì˜ˆì¸¡í•œ ìˆ«ìë³´ë‹¤ í¼")
  else:
    print("ì»´í“¨í„°ê°€ ì˜ˆì¸¡í•œ ìˆ«ìë³´ë‹¤ ì‘ìŒ")

    print(f'ë„ì „í•œ íšŸìˆ˜ {time}íšŒ')
    print(f'ì •ë‹µì€ {answer}')
```

```python
#ì‚¬ìš©ìê°€ ìƒê°í•œ ìˆ«ì ë§ì¶”ê¸°
print('1~100ì‚¬ì´ì˜ ìˆ«ì í•˜ë‚˜ë¥¼ ì •í•˜ì„¸ìš”')
print('ì´ ìˆ«ìë¥¼ ë§ì¶°ë³´ì„¸ìš”')

low = 1
high = 100
time = 0 #íšŸìˆ˜

while True:
  print(f'low : {low} | high : {high}')
  num = (low+high)//2
  answer = input(f'{num}ì¸ê°€? (ok / up / down)')
  time += 1

  if answer == "ok":
    print(f'ë„ì „í•œ íšŸìˆ˜ {time}íšŒ')
    print(f'ì •ë‹µì€ {num}')
    break
  elif answer == "up":
    low = num+1
  elif answer == "down":
    high = num-1
  else:
    print("ì˜ëª» ì…ë ¥ë¨")

```
