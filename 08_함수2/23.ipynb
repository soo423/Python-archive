### 함수의 scope
* 전역 변수: 모듈(파일) 내에서 접근할 수 있으며 모듈 내 어디에서든지 접근할 수 있다.
* 지역 변수: 함수 내에서 정의된 변수이며, 해당 함수 내에서만 접근 할 수 있다. 해당 함수의 실행이 종료되면 더 이상 함수 안의 지역 변수에 접근할 수 없다.

````python
#파이썬에서는 같은 변수 이름이 다른 값을 가질 수 있음
num = 41 #전역변수

def number():
  num = 3
  print(f'함수 안에서, num = {num}')

number()
print(f'global, num = {num}')
````

### 재귀 함수 : 자기 자신을 호출하는 함수

````python
def factorial(n):
  if n == 1:
    return 1
  else:
    return n * factorial(n-1)

print(factorial(3)) #3! = 3*2*1, 1! = 1
````

### 가변 인자

*args : 순차적인 입력에 대해서 개수에 상관없이 입력을 받을 수 있다.
```
def 함수(*args):
  실행코드

```

**kwargs : 키워드 인자에 대해서 어떤 이름의 인자 입력이든 자유롭게 입력 받을 수 있다.
```
def 함수(**kwargs):
  실행코드
```

func(*args, **kwargs) : 함수 호출시 어떤 형태의 입력이든 받을 수 있다.


````python
def mul(*args):
  result = 1
  for i in args:
    result *= 1
  return result

print(mul(3,4))
print(mul(5))
print(mul(2,2,2,2,2))
````

````python
def group_by_age(**kwargs):
  group = {'adult':[], 'non-adult':[]}
  for name, age in kwargs.items():
    if age > 18:
      group['adult'] += {name}
    else:
      group['non-adult'] += {name}
    return group

print(group_by_age())
print(group_by_age(kim=25, lee=16, park=30))
````

````python
def free_arg(*args, **kwargs):
  print(f'args = {args}')
  print(f'kwargs = {kwargs}')

free_arg()
free_arg(1,2,3)
free_arg(first = '1st', second='2nd')
free_arg(5, first='1st')

````

### 함수의 객체화
* 함수를 변수에 대입하고 다른함수의 인자로 입력


````python
def function(fnc):
  print(f'fnc(2,3) = {fnc(2,3)}')

def mul(a,b):
  return a*b

function(mul)
````

````python
def function():
  def new_function(a,b):
    return a*b
  return new_function

mul = function()
print(mul(2,3))
````

````python
def double(x):
  return 2*x

def func_with_print(fnc):
  def new_func(x):
    y = fnc(x)
    print(f'y={y}')
    return y
  return new_func

new_double = func_with_print(double)
print(new_double(5))
````

### 클로저 closure : 함수 안 지역 변수를 함수 밖에서도 간접적으로 사용할 수 있다.

````python
def function(a):
  def inner_function(b):
    return a*b
  return inner_function

double = function(2) #double = inner_function . a = 2
triple = function(3) #triple = inner_function . a = 3

print(double(10)) #inner_function(10). b = 10 외부 함수로부터 물려받은 a와 자신이 받은 인자 b를 곱한 값을 변환
print(triple(10)) #inner_function(10), b = 10, 자기가 정의된 구역 밖에서 호출되어도 자기를 만들어준 외부 함수의 변수 a에 접근할 수 있다 이를 자유변수(free variable)를 캡처한다고 표현함
````

````python
#매번 같은 머리말(로그 레벨, 이름)을 붙여야 할 때 클로저를 쓰면 편함
def make_logger(level):
  def log(message):
    print(f'[{level}]{message}')
  return log

info_logger = make_logger('INFO')
error_logger = make_logger('ERROR')

info_logger('서버가 시작되었습니다.') # 클로저는 함수가 종료된 후에도 내부의 지역 변수를 기억
info_logger('서버가 시작되었습니다.') # 전역변수 처럼 값이 유지
error_logger('서버가 종료됨')
````

* 시간 복잡도 : 입력의 크기에 따라 알고리즘이 실행되는 데에 필요한 시간의 양을 나타내는 개념. 빅오 표기법(Big O Notation)으로 표시,

```
O(n**2) 입력의 크기 n의 제곱에 비례하는 시간이 걸리는 복잡도 표현

```
