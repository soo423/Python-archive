### 문제6

```python
name_age = {"Kim": 10, "Lee": 15, "Park": 20, "Choi": 25}
adults = []
for k, v in name_age.items():
  if v >= 20:
    adults.append(k)
print(adults)
```

```python
name_age = {"Kim": 10, "Lee": 15, "Park": 20, "Choi": 25}
adults_name_age = dict()
for k, v in name_age.items():
  if v >= 20:
    adults_name_age[k] = v
print(adults_name_age)
```

```python
name_age = {"Kim": 10, "Lee": 15, "Park": 20, "Choi": 25}

adult_names = {"adult": [], "non-adult": []}

for k in name_age:
  if name_age[k] >= 20:
    adult_names["adult"].append(k)
else:
  adult_names["non-adult"].append(k)

print(adult_names)
```
