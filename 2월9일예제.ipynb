# 초 단위를 시분초로 변환



```python
seconds = int(input('초 입력 : '))
hours = seconds // 3600
seconds = seconds % 3600
minutes = seconds // 60
seconds = seconds % 60
print(f'{hours}시간 {minutes}분 {seconds}초')
```

# 주문 금액 계산기

```python
a = 2000
b = 3000
c = 2500

americano = int(input('아메리카노 주문 수량 입력'))
latte = int(input('라떼 주문 수량 입력'))
greentea = int(input('그린티 주문 수량 입력'))

sum = americano * a + latte * b + greentea * c
print(f'아메리카노 : {americano}개 / 가격 {americano * a}')
print(f'라떼 : {latte}개 / 가격 {latte * b}')
print(f'그린티 : {greentea}개 / 가격 {greentea * c}')
print(f'주문 금액 결과 : {sum}원')
```

```python
americano_price = 2000
latte_price = 3000
greentea_price = 2500

americano = int(input('아메리카노 주문 수량 입력'))
latte = int(input('라떼 주문 수량 입력'))
greentea = int(input('그린티 주문 수량 입력'))

sum = americano_price * americano + latte_price * latte + greentea_price * greentea
print(f'주문 금액 결과 : {sum}원')
```

# 몫과 나머지 구하기

```python
#무작위로 정수 두개(a,b) 생성한 다음에 a나누기b의 몫과 나머지를 출력
#a, b
#my_quoitent 몫
#my_remainder 나머지
#실제값과 예측값을 비교해서 출력함 : true, false

import random
a = random.randint(10, 50)
b = random.randint(2, 9)

your_quoitent = a // b
your_remainder = a % b

my_quoitent = int(input(f'{a}를 {b}로 나누면 몫은? '))
my_remainder = int(input(f'{a}를 {b}로 나누면 나머지는?'))

print(f'A 입력 결과 : {my_quoitent} || B 입력 결과 : {my_remainder}')
print(f'나눈 결과 {your_quoitent } || 나머지 {your_remainder}')
print(f'나눈 결과 {your_quoitent == my_quoitent} || 나머지 {your_remainder == my_remainder }')
```
