## 심화 예제

```python
#문제 6번(a)
n = 10
fibo = [1,1]
for i in range(n):
    num = fibo[-1] + fibo[-2]
    fibo = fibo + [num]
print(fibo)
```

```python
#문제7번
#1. i가 0일 때 0으로 나누려는 시도가 있었음
#2. d. 반복 과정에서 i=1일때 i로 나눈 나머지는 무조건 0이기 때문
#3. 해결 방법 : for i in range(n): > for i in range(2, n):으로 수정
n = 31
n_is_prime = True
for i in range(2, n):
  if n % i == 0:
    n_is_prime = False
print(n_is_prime)
```
